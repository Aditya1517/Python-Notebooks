```python
# broadcasting in numpy

a = np.tile(np.arange(0,40,10),(3,1))
print(a)
print("******************")
print(a.shape)
print("******************")
a=a.T
print(a)

# The given code uses NumPy to create and manipulate a 2D array. First, `np.arange(0, 40, 10)` generates a 1D array `[0, 10, 20, 30]`,
# which consists of values starting from 0 up to (but not including) 40 in steps of 10. This array is then replicated using
#`np.tile(..., (3, 1))`, which stacks the same row three times vertically to form a
# 3x4 matrix: `[[0, 10, 20, 30], [0, 10, 20, 30], [0, 10, 20, 30]]`. The shape of this array is printed as `(3, 4)`,
# meaning it has 3 rows and 4 columns. Then, the `.T` operation transposes the matrix, swapping rows with columns. As a result,
#the transposed matrix becomes `[[0, 0, 0], [10, 10, 10], [20, 20, 20], [30, 30, 30]]`, changing its shape to `(4, 3)`.
#This demonstrates how array tiling and transposing can be used to restructure data layouts effectively using NumPy.
```

```python
# adding a new axix i.e converting it to a 2D array
a = a[:,np.newaxis]
a.shape
```
