```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title('MyPlot')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("numbers")
plt.ylabel("squares")
plt.grid() # grid is visible here and in the above cell it is not visible
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') # here ro means color is red and o mean put dot the points
plt.grid()
plt.show()
```

```python
t = np.arange(0.,5.,2.) # values will go like 0,0.2,0.4,-------------,4.8 and 5 will be excluded

# blue dashes, red squares and green triangles
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.grid()
plt.legend() # this will be hepls to understand what line belongs to what here value like ^2 and ^2.2 and ^2.5 with the color will come in legend
plt.show()
```

```python
# controlling line properties
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=5.0)
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)

# use keywords args
lines[0].set(color='r', linewidth=2.0)

# or matlab style string value pairs
lines[1].set(color='g', linewidth=2.0)
plt.grid()
plt.show()
```

```python
# working with multiple figures and axes

def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.2)

plt.figure(1)
# subplot() command specifies numrows,numcols,fignums where fignum ranges from 1 to numrows*numcols
plt.subplot(211) # this means subplot of 2 roes and 1 column and fig number is 1
plt.grid()
plt.plot(t1,f(t1),'b-')

plt.subplot(212) # this means subplot of 2 roes and 1 column and fig number is 1
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1) # this is how we to return to the previous figure and can make changes here
plt.subplot(211)
plt.title('Easy as 1,2,3')

plt.show()
```
