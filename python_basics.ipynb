```python
# functions
def add(a,b):
  return a+b

print(add(2,3))

def add_sub(a,b):
  """
  this function adds and substracts two numbers
  """
  # above is the doc string
  return a+b,a-b

print(add_sub(2,3))
print(add_sub.__doc__)

# HCF
def computeHCF(a,b):
  smaller = b if a>b else a
  hcf = 1
  for i in range(1,smaller+1):
    if a%i == 0 and b%i == 0:
      hcf = i
  return hcf

print("hcf is "+str(computeHCF(98,78)))
```

```python
# # built in functions :- below are the some of the build in functions, check their usage
# abs()
# all()
# dir()
# divmod()
# enumerate()
# filter()
# isinstance()
# map()
# reduce()
# # map, filter, and reduce are used more in data science

def greet(name,msg):
  print("Hello {0} , {1}".format(name,msg))

greet("Satish","Good Morning")
# default argument concept
# kwargs - keyword arguments : **kwargs
# arbitrary arguments : * args

# recursive fucntions

#lambda functinos :- extensivly used in ml and ds
double = lambda x:x*2
print(double(5))

lst = [1,2,3,4,5]
even_lst = list(filter(lambda x:(x%2==0),lst))
print(even_lst)
# This Python code uses the filter() function along with a lambda expression to extract even numbers from a list. The lambda function lambda x: (x % 2 == 0) checks whether each element in the list lst is even by returning True for even numbers and False for odd ones. The filter() function applies this condition to all elements in the list, and only the elements that satisfy the condition (i.e., return True) are kept. Finally, list() is used to convert the filtered result into a list, resulting in even_lst containing only the even numbers [2, 4].


from functools import reduce

# Original list
numbers = [1, 2, 3, 4, 5, 6]

# 1. Using map() to square each number
squares = list(map(lambda x: x ** 2, numbers))
print("Squares:", squares)  # Output: [1, 4, 9, 16, 25, 36]

# 2. Using filter() to get even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", evens)  # Output: [2, 4, 6]

# 3. Using reduce() to get the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print("Product:", product)  # Output: 720


```
